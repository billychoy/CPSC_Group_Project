# COMBINE THE DATASETS

Finally, we are now in a position to combine all of these datasets. Before starting the merge, let's quickly visualize the number of rows in each of the cleaned datasets. This will help us decide which of the four merge strategies (left, right, inner, or outer) we should be using.

We have to merge three set of data together:

- combining two set of happiness data - happy_2016 + happy_2016
- combining useful Olympic Game data - countries + medals (aggregated from athletes)
- combining result in 1 and result in 2 together

We will discuss the methodology got joining them together in the follow one by one.

### 1. Combining two set of happiness data

After cleaning the happiness data, we could check for the row count of the key (country) from both table first.

```python
import seaborn as sns
import matplotlib.pyplot as plt

table = ['Happy 2016','Happy 2017']
unique_key = [happy_2016['Country'].nunique(),happy_2017['Country'].nunique()]
plots = sns.barplot(x=table, y=unique_key)
for bar in plots.patches:
    plots.annotate(bar.get_height(),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=14, xytext=(0, -10),
                   textcoords='offset points')
plt.xlabel("Tables", size=12)
plt.ylabel("Unique Key Count", size=12)
plt.title("Unique Key Count in Happiness Tables", size=18)
plt.show()
```

The key value count of both tables are not matched, therefore we could take a closer look for the unmatched rows before merging them together.

```python
set(happy_2016['Country'].unique())-set(happy_2017['Country'].unique())
```

```python
set(happy_2017['Country'].unique())-set(happy_2016['Country'].unique())
```

From the result of above checking, we notice that there are unmatched countries in both tables. The unmatched columns are around __5% of the total row counts__. We would decide to keep all data first, so that we can try to match the Olympic Game data in later stage. Therefore, we will merge the happinese data "happy_2016" and "happy_2016" into "happy_compare_clean" by using outer join to keep all the data first.

```python
happy_compare_clean = pd.merge(happy_2016, happy_2017, how='outer', on='Country', suffixes=('_2016', '_2017'))
##check any duplicated rows
len(happy_compare_clean) - happy_compare_clean['Country'].nunique()
```

### 2. Combining useful Olympic Game data

Although there are three tables in the sources data of Olympic Game, "events" is not related in our studies. Therefore, we would only consider the data about "athletes" and "countries". From the cleaning process, we aggregated the "athletes" data into "medals", so we woukd try to combine the "medals" and "countries" table.

First of all, we should define the key in both tables. In this case, the key are of two tables are not in the same name. The key of two tables are as follow:
- Medals table -> nationality
- Countries table -> code

```python
table = ['Medals','Countries']
unique_key = [medals['nationality'].nunique(),countries['code'].nunique()]
plots = sns.barplot(x=table, y=unique_key)
for bar in plots.patches:
    plots.annotate(bar.get_height(),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=14, xytext=(0, -10),
                   textcoords='offset points')
plt.xlabel("Tables", size=12)
plt.ylabel("Unique Key Count", size=12)
plt.title("Unique Key Count in Olympic Game Tables", size=18)
plt.show()
```

The key value count of both tables are not matched, therefore we could take a closer look for the unmatched rows before merging them together.

```python
set(medals['nationality'].unique())-set(countries['code'].unique())
```

```python
set(countries['code'].unique())-set(medals['nationality'].unique())
```

From the result of above checking, we notice that there are unmatched countries in both tables. The unmatched columns are around __8% of the total row counts__. However, we do would only consider the countries with medals, the countries are not appearing in medals table can be ignored. Therefore, we will merge the Olympic data "medals" and "countries" into "olympic" by using left join to keep all the medal related data first.

```python
olympic = pd.merge(medals, countries, how='left', left_on='nationality', right_on='code')
##check any duplicated rows
len(olympic) - olympic['nationality'].nunique()
```
